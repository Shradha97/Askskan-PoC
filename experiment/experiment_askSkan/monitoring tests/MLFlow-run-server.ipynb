```python
import yaml
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env 

```

```python
# Load the YAML file
file_path = '/Users/saishradhamohanty/Desktop/Repo/Prototypes/askskan/monitoring/mlflow_config.yaml'
with open(file_path, 'r') as file:
    data = yaml.safe_load(file)

# Define the keys to add
additional_keys = {
    'openai_api_type': "azuread",
    'openai_api_key': os.getenv("AZURE_OPENAI_KEY"),
    'openai_deployment_name': os.getenv("AZURE_OPENAI_DNAME"),
    'openai_api_base': os.getenv("AZURE_OPENAI_ENDPOINT"),
    'openai_api_version': "2023-05-15"
}

# Update the YAML data with the additional keys
if 'routes' in data and isinstance(data['routes'], list):
    for route in data['routes']:
        if 'model' in route and 'config' in route['model']:
            if route['model']['config'] is None:
                route['model']['config'] = additional_keys.copy()
            else:
                route['model']['config'].update(additional_keys)


# Write the updated data back to the YAML file
with open(file_path, 'w') as file:
    yaml.dump(data, file)
```
